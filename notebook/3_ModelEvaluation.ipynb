# Phase 3: Model Training, Evaluation & Comparison
## Student Graduation Prediction System - Research Project

**Objective**: Train multiple ML algorithms, optimize hyperparameters, perform comprehensive evaluation, dan compare performance.

**Models to Compare**:
1. Logistic Regression (Baseline)
2. Decision Tree & Random Forest (Ensemble)
3. Gradient Boosting & XGBoost (Advanced Ensemble)
4. SVM (Support Vector Machine)
5. Neural Network (Deep Learning)

**Evaluation Methods**:
- Train/Test Split dengan stratification
- K-Fold Cross-Validation
- Hyperparameter tuning (Grid Search)
- ROC-AUC, Precision, Recall, F1-Score
- Confusion Matrix & Error Analysis
- Visualization & interpretability

## 1. Setup & Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, roc_curve, auc, confusion_matrix, classification_report)
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# Try to import xgboost if available
try:
    import xgboost as xgb
    HAS_XGBOOST = True
except:
    HAS_XGBOOST = False

# Load data
print("Loading engineered features...")
X = pd.read_csv('features_engineered_scaled.csv')
df = pd.read_csv('../data/dataset.csv')
y = df['lulus_tepat_waktu'].copy()

# Encode target
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"[OK] Data loaded:")
print(f"  - X shape: {X.shape}")
print(f"  - y shape: {y_encoded.shape}")
print(f"  - Features: {list(X.columns)[:5]}...")

# Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\n[OK] Data split:")
print(f"  - Train: {X_train.shape[0]} samples")
print(f"  - Test: {X_test.shape[0]} samples")
print(f"  - Train class distribution: {np.bincount(y_train)}")
print(f"  - Test class distribution: {np.bincount(y_test)}")
```

## 2. Model Training: Baseline Models

```python
print("=" * 70)
print("BASELINE MODEL TRAINING")
print("=" * 70)

models = {}
results = []

# Model 1: Logistic Regression
print("\n[1] Training Logistic Regression...")
models['LogisticRegression'] = LogisticRegression(random_state=42, max_iter=1000)
models['LogisticRegression'].fit(X_train, y_train)
y_pred_lr = models['LogisticRegression'].predict(X_test)
y_proba_lr = models['LogisticRegression'].predict_proba(X_test)[:, 1]
acc_lr = accuracy_score(y_test, y_pred_lr)
print(f"   [OK] Logistic Regression - Accuracy: {acc_lr:.4f}")
results.append(('Logistic Regression', acc_lr, y_pred_lr, y_proba_lr, models['LogisticRegression']))

# Model 2: Decision Tree
print("\n[2] Training Decision Tree...")
models['DecisionTree'] = DecisionTreeClassifier(max_depth=5, random_state=42)
models['DecisionTree'].fit(X_train, y_train)
y_pred_dt = models['DecisionTree'].predict(X_test)
y_proba_dt = models['DecisionTree'].predict_proba(X_test)[:, 1]
acc_dt = accuracy_score(y_test, y_pred_dt)
print(f"   [OK] Decision Tree - Accuracy: {acc_dt:.4f}")
results.append(('Decision Tree', acc_dt, y_pred_dt, y_proba_dt, models['DecisionTree']))

# Model 3: Random Forest
print("\n[3] Training Random Forest...")
models['RandomForest'] = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
models['RandomForest'].fit(X_train, y_train)
y_pred_rf = models['RandomForest'].predict(X_test)
y_proba_rf = models['RandomForest'].predict_proba(X_test)[:, 1]
acc_rf = accuracy_score(y_test, y_pred_rf)
print(f"   [OK] Random Forest - Accuracy: {acc_rf:.4f}")
results.append(('Random Forest', acc_rf, y_pred_rf, y_proba_rf, models['RandomForest']))

# Model 4: Gradient Boosting
print("\n[4] Training Gradient Boosting...")
models['GradientBoosting'] = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
models['GradientBoosting'].fit(X_train, y_train)
y_pred_gb = models['GradientBoosting'].predict(X_test)
y_proba_gb = models['GradientBoosting'].predict_proba(X_test)[:, 1]
acc_gb = accuracy_score(y_test, y_pred_gb)
print(f"   [OK] Gradient Boosting - Accuracy: {acc_gb:.4f}")
results.append(('Gradient Boosting', acc_gb, y_pred_gb, y_proba_gb, models['GradientBoosting']))

# Model 5: SVM
print("\n[5] Training SVM...")
models['SVM'] = SVC(kernel='rbf', probability=True, random_state=42)
models['SVM'].fit(X_train, y_train)
y_pred_svm = models['SVM'].predict(X_test)
y_proba_svm = models['SVM'].predict_proba(X_test)[:, 1]
acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"   [OK] SVM - Accuracy: {acc_svm:.4f}")
results.append(('SVM', acc_svm, y_pred_svm, y_proba_svm, models['SVM']))

# Model 6: Neural Network
print("\n[6] Training Neural Network...")
models['NeuralNetwork'] = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=42)
models['NeuralNetwork'].fit(X_train, y_train)
y_pred_nn = models['NeuralNetwork'].predict(X_test)
y_proba_nn = models['NeuralNetwork'].predict_proba(X_test)[:, 1]
acc_nn = accuracy_score(y_test, y_pred_nn)
print(f"   [OK] Neural Network - Accuracy: {acc_nn:.4f}")
results.append(('Neural Network', acc_nn, y_pred_nn, y_proba_nn, models['NeuralNetwork']))

print("\n" + "=" * 70)
print("[OK] All baseline models trained successfully!")
```

## 3. Comprehensive Model Evaluation

```python
print("=" * 70)
print("COMPREHENSIVE MODEL EVALUATION")
print("=" * 70)

# Calculate all metrics for each model
evaluation_metrics = []

for name, acc, y_pred, y_proba, model in results:
    print(f"\n {name}:")
    
    # Binary metrics
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1_weighted')
    cv_mean = cv_scores.mean()
    cv_std = cv_scores.std()
    
    print(f"   Accuracy:   {acc:.4f}")
    print(f"   Precision:  {precision:.4f}")
    print(f"   Recall:     {recall:.4f}")
    print(f"   F1-Score:   {f1:.4f}")
    print(f"   ROC-AUC:    {roc_auc:.4f}")
    print(f"   CV Score:   {cv_mean:.4f} (+/- {cv_std:.4f})")
    
    evaluation_metrics.append({
        'Model': name,
        'Accuracy': acc,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc,
        'CV-Score': cv_mean,
        'CV-Std': cv_std
    })

# Create comparison dataframe
eval_df = pd.DataFrame(evaluation_metrics)
print("\n" + "=" * 70)
print("MODEL PERFORMANCE COMPARISON")
print("=" * 70)
print(eval_df.round(4).to_string(index=False))

# Best model
best_model_idx = eval_df['F1-Score'].idxmax()
best_model = eval_df.iloc[best_model_idx]['Model']
print(f"\n Best Model by F1-Score: {best_model}")
print(f"   F1-Score: {eval_df.iloc[best_model_idx]['F1-Score']:.4f}")
```

## 4. Hyperparameter Tuning (Grid Search)

```python
print("=" * 70)
print("HYPERPARAMETER TUNING - RANDOM FOREST (BEST BASELINE)")
print("=" * 70)

# Grid search parameters for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

print(f"\nParameter Grid:")
print(f"  n_estimators: {param_grid['n_estimators']}")
print(f"  max_depth: {param_grid['max_depth']}")
print(f"  min_samples_split: {param_grid['min_samples_split']}")
print(f"  min_samples_leaf: {param_grid['min_samples_leaf']}")

print(f"\nRunning Grid Search CV (5-fold)...")
gs = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, n_jobs=-1, scoring='f1')
gs.fit(X_train, y_train)

print(f"\n✓ Grid Search completed!")
print(f"\nBest Parameters:")
for param, value in gs.best_params_.items():
    print(f"  {param}: {value}")

print(f"\nBest CV Score: {gs.best_score_:.4f}")

# Evaluate optimized model
y_pred_best = gs.best_estimator_.predict(X_test)
y_proba_best = gs.best_estimator_.predict_proba(X_test)[:, 1]

print(f"\nOptimized Model Performance on Test Set:")
print(f"  Accuracy:  {accuracy_score(y_test, y_pred_best):.4f}")
print(f"  Precision: {precision_score(y_test, y_pred_best):.4f}")
print(f"  Recall:    {recall_score(y_test, y_pred_best):.4f}")
print(f"  F1-Score:  {f1_score(y_test, y_pred_best):.4f}")
print(f"  ROC-AUC:   {roc_auc_score(y_test, y_proba_best):.4f}")

# Save best model
import joblib
joblib.dump(gs.best_estimator_, 'model_random_forest_optimized.pkl')
print(f"\n✓ Optimized model saved as 'model_random_forest_optimized.pkl'")
```

## 5. Error Analysis & Classification Report

```python
print("=" * 70)
print("ERROR ANALYSIS - OPTIMIZED RANDOM FOREST")
print("=" * 70)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_best)
print("\nConfusion Matrix:")
print(f"  {'':15s} {'Predict: Tidak':>15s} {'Predict: Lulus':>15s}")
print(f"  {'Actual: Tidak':15s} {cm[0, 0]:15d} {cm[0, 1]:15d}")
print(f"  {'Actual: Lulus':15s} {cm[1, 0]:15d} {cm[1, 1]:15d}")

tn, fp, fn, tp = cm.ravel()
print(f"\nDetailed Metrics:")
print(f"  True Negatives (TN): {tn}")
print(f"  False Positives (FP): {fp}") 
print(f"  False Negatives (FN): {fn}")
print(f"  True Positives (TP): {tp}")

print(f"\nSpecificity: {tn / (tn + fp):.4f}" if (tn + fp) > 0 else "")
print(f"Sensitivity (Recall): {tp / (tp + fn):.4f}" if (tp + fn) > 0 else "")

# Classification Report
print("\n" + "=" * 70)
print("CLASSIFICATION REPORT")
print("=" * 70)
print(classification_report(y_test, y_pred_best, target_names=label_encoder.classes_, digits=4))

# Misclassification analysis
print("\n" + "=" * 70)
print("MISCLASSIFICATION ANALYSIS")
print("=" * 70)

misclassified_indices = np.where(y_pred_best != y_test)[0]
if len(misclassified_indices) > 0:
    print(f"\nTotal misclassified: {len(misclassified_indices)} / {len(y_test)}")
    print(f"Misclassification rate: {len(misclassified_indices) / len(y_test) * 100:.2f}%")
    
    print(f"\nMisclassified samples:")
    for idx in misclassified_indices[:5]:  # Show first 5
        actual = label_encoder.inverse_transform([y_test[idx]])[0]
        predicted = label_encoder.inverse_transform([y_pred_best[idx]])[0]
        confidence = y_proba_best[idx] if y_pred_best[idx] == 1 else 1 - y_proba_best[idx]
        print(f"  Actual: {actual:10s}, Predicted: {predicted:10s}, Confidence: {confidence:.4f}")
else:
    print("\n✓ No misclassifications! Perfect predictions on test set.")
```
